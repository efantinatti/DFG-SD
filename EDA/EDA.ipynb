```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#1. Load the Dataset
df = pd.read_csv('../Dataset/education-economy-data.csv')
```

```python
# Display first few rows and data info to check structure
print("Dataset Info:")
print(df.info())
display(df.head())
```

```python
# 2. Descriptive Statistics
# -------------------------
desc_stats = df.describe()
print("\nDescriptive Statistics:")
display(desc_stats)

# Set plotting style
sns.set_theme(style="whitegrid")
```

```python
 #3. Visualizations

# Figure A: Distributions of Key Indicators
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Literacy Rate Distribution
sns.histplot(df['Literacy Rate'], kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Literacy Rate')

# GDP per Capita Distribution
sns.histplot(df['GDP per Capita (Current USD)'], kde=True, ax=axes[1], color='salmon')
axes[1].set_title('Distribution of GDP per Capita')

# Unemployment Rate Distribution
sns.histplot(df['Unemployment Rate (%)'], kde=True, ax=axes[2], color='green')
axes[2].set_title('Distribution of Unemployment Rate')

plt.tight_layout()
plt.show()

# Figure B: Correlation Heatmap
plt.figure(figsize=(10, 8))
# Select only numeric columns for correlation matrix
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Economic & Social Indicators')
plt.tight_layout()
plt.show()

# Figure C: Literacy Rate vs. GDP per Capita (by Continent)
plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df, 
    x='Literacy Rate', 
    y='GDP per Capita (Current USD)', 
    hue='Continent', 
    palette='viridis', 
    s=100, 
    alpha=0.8
)
plt.title('Literacy Rate vs. GDP per Capita by Continent')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()

# Figure D: Unemployment Rate by GDP Category
plt.figure(figsize=(10, 6))
# Define order for the categorical variable
category_order = ['Very Low', 'Low', 'Mid', 'High']

sns.boxplot(
    data=df, 
    x='GDP per Capita Category', 
    y='Unemployment Rate (%)', 
    order=category_order, 
    palette="Set2"
)
plt.title('Unemployment Rate by GDP per Capita Category')
plt.tight_layout()
plt.show()
```
